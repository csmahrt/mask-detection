# **Downloading Images from Bing**



Since scraping Google Images appears to be broken at the time of starting this project, I've opted to scrape Bing instead. I'm using the bing-image-downloader library which operates similiarly to how the google-image-downloader library used to work.

I chose 17 phrases that resulted in decent results on Bing. These phrases include masked and unmasked people along with a few random animals and backgrounds. I chose these non-pertinent images in order to have the classifier work whether there's a human in the image. I also used phrases that returned images of people that covered their nose and/or mouth with something other than a mask or bandana in order to avoid training a nose/mouth detector.

```python
from bing_image_downloader import downloader
```
